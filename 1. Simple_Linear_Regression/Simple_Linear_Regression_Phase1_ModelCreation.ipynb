```python
#importing the pandas library and giving it an alias name pd. Pandas library helps to work with data.
import pandas as pd
```

```python
"""
read_csv() - a function from pandas, and it is used to read a CSV file. It loads the file's content into a pandas DataFrame and stores it in a variable
called dataset.
"""
dataset = pd.read_csv("Salary_Data.csv")
```

```python
#display the contents of the dataset variable - which is pandas dataframe.
dataset
```

```python
#Input/Output data split
"""
To access specific columns or a column from a dataset variable (dataframe). Using double brackets [[]] to return as a dataframe.
Then, we store it in a variable called independent. We named this variable "independent" because it is an input(X).
"""
independent = dataset[["YearsExperience"]]
#Display the contents of the independent variable
independent
```

```python
#Input/Output data split
"""
To access specific columns or a column from a dataset variable (dataframe). Using double brackets [[]] to return as a dataframe.
Then, we store it in a variable called dependent. We named this variable "dependent" because it is an output(y).
"""
dependent = dataset[["Salary"]]
#Display the contents of the dependent variable
dependent
```

```python
#Split train and test data
#importing the train_test_split() function from sklearn.model_selection module. This function splits the dataset into training and testing sets.
"""
call train_test_split() and pass four arguments - independent variable(input), dependent variable(output), test_size as 0.30 (30% of data for testing0),
70% data for training, and then value of random_state is 0 to ensure same split every time for reproducibility.
Assign these to 4 variables X_train - 70% (YearsExperience) for training, X_test - 30% (YearsExperience) for testing, y_train - 70% (Salary) for training,
y_test - 30% (Salary) for testing.
"""
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(independent, dependent, test_size = 0.30, random_state = 0)
```

```python
#Model Creation - weight and intercept.
#import the LinearRegression class from sklearn.linear_model module.
#Create an instance (object) of the LinearRegression class from Scikit-Learn.
#This object, named 'regressor', will be used to train and apply a linear regression model.
#Regressor is the specific object that holds the trained model once fitted.
#fit() function is used to train a machine learning model. It takes training data as arguments.
#Find the best-fitting line(by calculating slope and intercept) that minimizes the error between predicted and actual values.
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
#regressor.coef_ returns the weight(coefficient) learned by the linear regression model during training.
weight = regressor.coef_
weight
```

```python
#regressor.intercept_ returns the intercept(bias) of the regression line. The value of the output(Salary) when the input(YearExperience) is zero.
bias = regressor.intercept_
bias
```

```python
#Evaluation metrics
#Use the trained linear regression model to predict the output(Salary) for the test input data(X_test).
#The predicted results are stored in a variable y_pred.
y_pred = regressor.predict(X_test)
```

```python
#Calculate R² score score
#import r2_score function from sklearn.metrics module to evaluate model performance.
#Pass the actual output and predicted output as arguments to r2_score function. Calculate the R² score and store it in a variable class r_score.
#R² score measures how well the model's prediction matches the actual test values.
#A score closer to 1 indicates a better fit.
from sklearn.metrics import r2_score
r_score = r2_score(y_test, y_pred)
```

```python
#Display the value stored in the r_score variable. It is closer to 1, which means this model fits the data very well.
r_score
```

```python
#import the pickle module to save Python objects into a file.
#Define the file name with extension .sav where the trained model to be stored.
import pickle
filename = "finalized_model_linear.sav"
```

```python
#Opens the file whose name is stored in the filename variable in write binary mode.
#pickle.dump function serializes(converts) the regressor object into a binary format and then writes the data into a file.
pickle.dump(regressor, open(filename, 'wb'))
```

```python
#Opens the saved file in a read binary mode. 
#pickle.load() deserializes(converts) from binary format back into python object.
#The variable loaded_model now holds the exact model that we trained.
#loaded_model to predict the salary based on input(experience) and stored it in a variable result.
loaded_model = pickle.load(open("finalized_model_linear.sav", 'rb'))
result = loaded_model.predict([[13]])
```

```python
#Display the value stored in a variable result.
result
```
